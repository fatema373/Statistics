## Exponential Distribution
- The number of miles that a particular car can run before its battery wears out is exponentially distributed with an average of 10,000 miles. The owner of the car needs to take a 5000-mile trip.
- Using <b>scipy.stats.expon()</b> and <b>numpy.random.exponential()</b>
    - What is the probability that he will be able to complete the trip without having to replace the car battery?
    - What is the probability of the battery failure during the trip?
    - Plot the probability distribution in both cases.
    - Calculate the mean and variance.
    - <b>What if the car already spent 2000 miles</b>. In this case, What will be the probability that he will be able to complete the trip without having to replace the car battery?

### Using SciPy

```python
from scipy.stats import expon
```

```python
lambda_= 10000
x=np.arange(0,50000)
pdf = st.expon.pdf(x,scale=lambda_,loc=0)
```

```python
sns.lineplot(x=x,y=pdf)
print(expon.mean(scale = lambda_))
print(expon.var(scale = lambda_))
```

```python
x = 5000
probability_of_complition = 1 - expon.cdf(x = 5000,scale = lambda_)
print("The probability that he will be able to complete the trip without having to replace the car battery:")
probability_of_complition
```

```python
x = 5000
probability_of_failure = expon.cdf(x = 5000,scale = lambda_)
print(" the probability of the battery failure during the trip:")
probability_of_failure
```

```python
#Or
1-probability_of_complition
```

### If the car already spent 2000 miles.

```python
remaining_dist= x + 2000
probability= 1-expon.cdf(x = remaining_dist,scale = lambda_)
print("The probability that he will be able to complete the trip without having to replace the car battery:")
probability
```

```python
#or
probability= 1-expon.cdf(x = x,scale = lambda_,loc=-2000)
print("The probability that he will be able to complete the trip without having to replace the car battery:")
probability
```
